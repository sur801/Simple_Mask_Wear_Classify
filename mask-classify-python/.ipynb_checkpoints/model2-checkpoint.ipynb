```python
train_x=np.array(train_x).reshape(-1,50,50,1)
print(train_x.shape)
#x=tf.keras.utils.normalize(x,axis=1)
train_x = train_x * 1 / 255.0 # normailze 0~1 
train_x = np.float32(train_x) # make float32
print(train_y)
print(train_y.shape)
train_y = tf.keras.utils.to_categorical(train_y)

train_y = np.array(train_y).reshape(-1,2)
print(train_y)
print(train_y.shape)
```

```python
def weight_variables(shape):
  initial = tf.truncated_normal(shape, stddev=0.1)
  return tf.Variable(initial)

def bias_variable(shape):
  initial = tf.constant(0.1, shape=shape)
  return tf.Variable(initial)

def conv2d(x, W, stride):
  return tf.nn.conv2d(x, W, strides=stride, padding='VALID')

def maxpool(x):
    return tf.nn.max_pool(x, ksize = [1, 2, 2, 1] , strides = [1, 1, 1, 1], padding = 'VALID')

def mask_convnet(x):
    # input shape (50, 50, 1)
    
    #1 layer
    W_conv1 = weight_variables([3,3,1,100])
    b_conv1 =  bias_variable([100])
    h_conv1 = tf.nn.relu(conv2d(x, W_conv1, [1,2,2,1]) + b_conv1)
    Max1 = maxpool(h_conv1)
    #h_conv1 = tf.nn.conv2d(x, W_conv1, strides = [1, 2, 2, 1], padding = 'VALID') #stride = 2
    #Relu1 = tf.nn.relu(Conv1 + b_conv1)
    #Max1 = tf.nn.max_pool(Relu1, ksize = [1, 2, 2, 1] , strides = [1, 1, 1, 1], padding = 'VALID') #stride = 2

    #2 layer
    W_conv2 = weight_variables([3, 3, 100, 64])
    b_conv2 = bias_variable([64])
    h_conv2 = tf.nn.relu(conv2d(Max1, W_conv2, [1,1,1,1]) + b_conv2)
    #Conv2 = tf.nn.conv2d(Max1, W_conv2 , strides = [1, 1, 1, 1] ,padding = 'VALID') #stride = 2
    #Relu2 = tf.nn.relu(Conv2 + bias_variable([64]))
    #Max2 = tf.nn.max_pool(Relu2, ksize = [1, 2, 2, 1] , strides = [1, 1, 1, 1], padding = 'VALID')
    Max2 = maxpool(h_conv2)
    #print(Max2.shape) #>> (83, 20, 20, 64)

    #3 layer fully connected layer
    Flat = tf.reshape(Max2, [-1, 20*20*64]) # flatten
    #print(Flat.shape) #>> (83, 25600)
    Fc3 = tf.nn.relu(tf.matmul(Flat, weight_variables([20*20*64,50] )) + bias_variable([50]))
    Norm = tf.nn.dropout(Fc3,0.2)
    #print(Norm.shape)
    #4 layer
    #Fc4 = tf.layers.Dense(2, activation='softmax')
    Fc4 = tf.nn.softmax(tf.matmul(Norm, weight_variables([50,2] )) + bias_variable([2]))
    return Fc4
    #print(Fc4.shape) (5749, 2)
```

```python
def train_model(train_data) :
    train_epoch = 30
    x = tf.placeholder(tf.float32, shape = [None, 50,50,1])
    #x = train_data
    #y = tf.placeholder(tf.float32, shape = [None, 2])
    y_ =tf.placeholder(tf.float32, shape = [None, 2])

    #print(train_x)
    y_out = mask_convnet(x)    
    cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y_, logits=y_out))
    
    train_step = tf.train.AdamOptimizer(1e-3).minimize(cross_entropy)
    correct_prediction = tf.equal(tf.argmax(y_out, 1), tf.argmax(y_, 1))
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
    # use adam optimizer
    
    with tf.Session() as sess:
        saver = tf.train.Saver()
        sess.run(tf.global_variables_initializer())
        for step in range(train_epoch+1):
            train_accuracy = accuracy.eval(feed_dict={x: train_data, y_: train_y})
            print("epoch : %d, training accuracy : %g"% (step, train_accuracy))
            train_step.run(feed_dict={x: train_data, y_:train_y})

train_model(train_x)
```
